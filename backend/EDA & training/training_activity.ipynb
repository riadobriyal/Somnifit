```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score
%matplotlib inline 
```

```python
df = pd.read_csv('../datasets/activity_dataset.csv')
df.head()
```

```python
x = df.drop(df.columns[6],axis=1)
y = df['Exercise Recommendation Plan'] #output
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer

scaler = StandardScaler() #normaliser for numbers 
ohe_gender = OneHotEncoder(drop='if_binary') #changing gender to 0 and 1
ohe_BMIcase= OneHotEncoder()


ct = ColumnTransformer([
    ("StandardScaler", StandardScaler(), [0, 1, 2, 4]), #2in1 
    ("OneHotEncode_gender", ohe_gender , [3]),
    ("OneHotEncode_BMIcase", ohe_BMIcase, [5])
])

x_train = ct.fit_transform(x_train)  #applying the normalisation
x_test = ct.transform(x_test)  
```

```python
x_train
```

```python

log_model = LogisticRegression()
```

```python
log_model.fit(x_train,y_train)
```

```python
y_pred = log_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)) #telling us model is bad not good 
```

```python
import pickle
pickle.dump(log_model,open('../models/log_reg_activity.pkl','wb'))
pickle.dump(ct, open('../models/column_transformer_activity.pkl','wb'))
```
